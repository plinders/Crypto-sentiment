```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
parse_dates = ['time']
df = pd.read_csv(open("twitter-out.txt", "r"), names=['time', 'tweet', 'id', 'followers', 'score', 'weighted_score'], parse_dates=parse_dates)
```

```python
df.info()
```

```python
fig, ax = plt.subplots(figsize=(14,6))
ax.hist(df['followers'], alpha=0.5, label="Unfiltered", bins = 100000)
ax.hist(df[df['followers'] > 100]['followers'], alpha=0.5, label=">100", bins = 100000)
ax.set_xscale('log')




plt.show()
```
