```python
def twitter_setup():
    auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
    auth.set_access_token(ACCESS_TOKEN, ACCESS_SECRET)
    
    api = tweepy.API(auth)
    return api
```

```python
extractor = twitter_setup()

search_text = "$btc"

tweets = extractor.search(search_text, rpp=100)

print("Number of tweets extracted: {}.\n".format(len(tweets)))

print("5 recent tweets:\n")
for tweet in tweets[:2]:
    print(tweet.text)
    print()
```

```python
for tweet in tweepy.Cursor(extractor.search, q = search_text, rpp=100).items(100000):
    print(tweet.text)
    print()
```
